--OPERATORS IN PYTHON-
1.   ARITHEMATIC OPERATOR
2.   RELATIONAL OPERATOR
3.   LOGICAL OPERATOR
4.   BITWISE OPERATOR
5.   ASSIGNMENT OPERATOR
6.   MEMBERSHIP OPERATOR
















```python
#arithematic operator
print(5+2)
print(5-2)
print(7*4)
print(7/2)
print(7//2) #integer division
print(5%2)#modulus operator->gives remainder
print(2**3)#power of ->2^3=8
```

```python
#relational operator->used when we have to compare the operands
print(4>5)
print(4<5)
print(2>=1)
print(3<=2)
print(1==1)
print(2!=5)
```

```python
#logical operator
print(1 and 0) #for and operator both operands should be true then it is true else false
print(1 or 0) #for or operator one of the operand should be true
print(not 1)
```

```python
#bitwise operator
#biwise and
print(2 & 3) #010
             #011
             #---
             #010->2 in decimal
#bitwise or
print(2 | 3) #010
             #011
             #---
             #011->3 in decimal
#bitwise xor
print(2 ^ 3) #010
             #011
             #---
             #001->1in decimal #high is only when inputs are different or if number of 1's are odd
#bitwise not
print(~3) #~011=-100->-4 in decimal,it flips all the bits in binary including the sign bit
#bitwise left shift
print(2 << 3) #shift the bits of binary no. 2 by three positions
              #000 010-> 010 000->16 in decimal
#bitwise right shift
print(2>>3)    #shift the bits of binary 2 by three postions to right
               #binary 2->00000010
               #shift by 1->00000001->1 in decimal
               #shift by 2->00000000->0 in decimal
               #shift by 3->00000000->still 0
               #key rule for positive numbers a>>b=a//(2**b)
```

```python
#assignment operator
a=2 #simply assigning variable a value =2
print(a)
a%=2 #a=a%2
print(a)
```

```python
#membership operator
print('A' in 'Adil') # the membership in operator checks whether this is present in that
print(1 in [1,2,3,4,5,6])
print(1 not in [5,4,6,7])
```

If-else in python
