1.Creating a string

```python
x="adil"
y='adil'
z='''adil'''
w="""adil"""
print(x)
print(y)
print(z)
print(w)
```

2.Accessing a string

```python
s="adil rasheed"
print(s[0])#positive indexing
print(s[2])
print(s[-1])#negative indexing  -> -1 means last index of a string , and from last to first index goes as -1,-2,-3...
```

```python
#slicing
s="python programming"
print(s[0:5])#it means print the string from 0 to 5th index but 5th index is not included
print(s[6::-1]) #it means print from 6th to zeroth index with step 1 in reverse order
print(s[:])#it means print all the string
print(s[::-1])#it means print all the string in reverse order
```

4.Operations on strings

* Arithematic operations
* Relational operations
* Logical operations
* Loops on Strings
* Membership operations  




```python
#arithematic operations
print("jammu" + "kashmir")
print("kashmir"*5)
```

```python
#relational operations
"j&K"=="INDIA"
"jammu"=="jammu"
"delhi"!="mumbai"
"jammu"<="kashmir"
```

```python
#logical
"jammu" and "kashmir"
"jammu" or "kashmir"
"" and "kashmir" #empty string means false ,so false and true output is false
"" or "kashmir"#false or true output is true
```

```python
#loop on strings
for i in "jammu":
  print(i ,end=" ")
```
