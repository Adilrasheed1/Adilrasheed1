1.CREATING A TUPLE

```python
#empty
t1=()
print(t1)
#single element tuple
t2=(2,)
print(t2)
#homogeneous tuple
t3=(1,2,3,4,5)
print(t3)
#hetrogeneous tuple
t4=(1,2,3,True,[1,2,3])
print(t4)
```

2. Accessing items from tuple

*   Indexing
*   slicing



```python
#indexing
t1=(1,2,3,4)
#negative indexing
print(t1[-1])
#positive indexing
print(t1[0])
```

```python
#slicing
print(t1[:])
print(t1[::-1])
```

3. editing items of tuple

items of a tuple cannot be edited or we can not add items to a tuple because tuples are immutable

4.Deleting items

```python
t1=(1,2,3,4,5)
del t1
#this works but if we want to delete particular items of tuple that will not work
```

4. OPERATIONS ON TUPLES


*  ARITHEMATIC
*  MEMBERSHIP
*  LOOP




```python
#arithematic(+,*)
t1=(1,2,3)
t2=(4,5,6)
print(t1+t2)
print(t1*2)
```

```python
#membership
print(1 in t1)
print(2 not in t2)
print(4 not in t2)
```

```python
#loop
for i in t1:
  print(i)
```

5. TUPLE FUNCTIONS

    len/min/max/sum/sorted/count/index

```python
t1=(1,2,3,5,3,2,9,8,5,4,3,9,0)
print(len(t1))
print(min(t1))
print(max(t1))
print(sum(t1))
print(sorted(t1))
print(t1.count(3))
print(t1.index(3))
```

```python
#special syntax
#tuple unpacking
a,b,c=(1,2,3)
print(a,b,c)
```

```python
#swapping
a=1
b=2
a,b=b,a
print(a,b)
```

```python
#zipping tuples
a=(1,2,3,4)
b=(5,6,7,8)
tuple(zip(a,b))
```

**SETS**

A set is  an unordered collection of items. Every set element is unique(no duplicates) and must bet immutable however a set itself is mutable we can add or remove items from a set.

**CHARATERISTICS**



*   UNORDERED
*   MUTABLE
*   NO DUPLICATES
*   CAN'T CONTAIN MUTABLE DATA TYPES




CREATING A SET


```python
#empty
s=set()
print(s)
```

```python
#1d set
s1={1,2,3,4}
print(s1)
```

```python
#2d set
#s2={1,2,3,4,{8,9,7},} this throws error because set itsel is mutable but set cannot contain mutable elements
```

ACCESING ELEMENTS

```python
s1={1,2,3,4,5}
#we can not access tjhe elements of  a set because set is unordered
```

EDITING ITEMS

```python
#CANNOT EDIT ITEMS OF SET BECAUSE ELEMENTS OF SET ARE IMMUTABLE
```

ADDING ITEMS IN SETS

```python
s={1,2,3}
s.add(5)
print(s) #add function adds single element inside the set
s.update([6,7,8])
print(s) #update function adds multiple items inside the list
```

**SET OPERATIONS**

UNION(|)

INTERSECTION(&)

DIFFERENCE(-)

SYMMETRIC DIFFERENCE(^)

MEMBERSHIP TEST(in ,not in)
ITERATION


```python
#union
s1={1,2,3,4}
s2={3,4,5,6,7,8}
s1|s2
```

```python
#intersection
s1 & s2
```

```python
#difference
print(s1 - s2)
print(s2 - s1)
```

```python
#symmetric difference
s1 ^ s2
```

```python
#membership
print(1 in s1)
print(1 not in s1)
```

```python
#loop
for i in s1:
  print(i)
```

**FUNCTIONS**

```python
#len/min/max/sum/sorted
s={1,2,3,4,5,6,7,8}
print(len(s))
print(min(s))
print(max(s))
print(sum(s))
print(sorted(s))
```

```python
#union/update
s1={1,2,3,4}
s2={3,4,5,6,7,8}
s1.union(s2)
print(s1)
s1.update(s2)
print(s1)
```

**DICTIONARY**

dictionary in python is  a collection of key values used to store data values like a map ,which unlike other data types ,which hold only a single value as an element.

**CHARACTERISTICS**

*   MUTABLE
*   INDEXING HAS NO MEANING
*   KEYS CAN'T BE DUPLICATED
*   KEYS CAN'T BE MUTABLE ITEMS





CREATING A DICTIONARY

```python
#empty
d={}
print(d)
#1d
d1={'name':'adil','gender':'male'}
print(d1)
#2d
d2={'name':'adil','college':'ku','subjects':{
    'dsa':50,
    'math':49,
    'os':44
}}
print(d2)
```

ACCESING ITEMS OF DICT.


```python
mydict={'name':'adil','age':23}
print(mydict['name'])
print(mydict.get('age'))
```

```python
#adding key value pair
mydict['gender']='male'
print(mydict)
```

```python
#removing key value pair
mydict.pop('gender')
print(mydict)
```

```python
#EDITING KEY VALUE PAIR
mydict['age']=22
print(mydict)
```

 **DICTIONARY OPERATIONS**


*   MEMBERSHIP
*   ITERATION



```python
#membership
'name' in mydict
```

```python
#iteration
for i in mydict:
  print(i,mydict[i])
```

```python
#len/sorted/min/max
d1={'name':'adil','age':22,'gender':'male'}
len(d1)
```

```python
#min
min(d1)
```
